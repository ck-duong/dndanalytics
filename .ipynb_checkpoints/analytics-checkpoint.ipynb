```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
all_fp = os.path.join('data', 'dnd_chars_all.tsv')
alldf = pd.read_csv(all_fp, sep='\t')
alldf = alldf.drop(columns = ['name', 'ip', 'finger', 'hash'])
alldf.head()
```

```python
alldf.columns
```

```python
def clean_titles(word):
    if not pd.isna(word):
        listed = word.replace('-', ' ').split()
        cleaned = ' '.join(listed)
        cleaned = cleaned.title()
        return cleaned
    else:
        return word
```

```python
def clean_spells(word):
    if not pd.isna(word):
        spellbook = {}
        listed = word.split('|')
        
        for spell in listed:
            level = spell[-1]
            if level in spellbook.keys():
                spellbook[level].append(spell[:-2])
            else:
                spellbook[level] = [spell[:-2]]
                
        return spellbook
    return np.nan
```

```python
def clean_misc(col):
    def clean(word):
        if not pd.isna(word):
            return word.split('|')
        return np.nan
    return col.apply(clean)
```

```python
alldf['race'] = alldf['race'].apply(clean_titles)
alldf['background'] = alldf['background'].apply(clean_titles)
alldf['processedSpells'] = alldf['processedSpells'].apply(clean_spells)
alldf[['class', 'justClass', 'subclass', 'feats', 'skills', 'processedWeapons']] = \
alldf[['class', 'justClass', 'subclass', 'feats', 'skills', 'processedWeapons']].apply(clean_misc)
```
